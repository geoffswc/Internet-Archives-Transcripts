```python
#advertising query
df_advertising = pysqldf("""
SELECT * FROM df
WHERE subject LIKE '%advertising%'
AND year IS NOT NULL
""")
```

```python
#legal-related query
df_legal = pysqldf("""
SELECT * FROM df
WHERE subject LIKE '%legal%'
OR subject LIKE '%lawsuit%'
OR subject LIKE '%lawyer%'
OR subject LIKE '%litigation%'
OR subject LIKE '%law%'
AND year IS NOT NULL
""")
```

```python
df_transcripts = pd.read_csv('transcript_accuracy.csv') 
```

```python
#df_transcripts.head(1)
```

```python
advertising_transcripts = pysqldf("""
SELECT
    *
FROM
    df_advertising
JOIN
    df_transcripts
ON
    df_advertising.identifier = df_transcripts.file_name

""")
```

```python
len(advertising_transcripts)
```

```python
law_transcripts = pysqldf("""
SELECT
    *
FROM
    df_legal
JOIN
    df_transcripts
ON
    df_legal.identifier = df_transcripts.file_name

""")
```

```python
len(law_transcripts)
```
