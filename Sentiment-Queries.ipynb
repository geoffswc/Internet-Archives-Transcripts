```python
import pandas as pd
from pandasql import sqldf
pysqldf = lambda q: sqldf(q, globals())
```

```python
df_sentiment_scores = pysqldf(
"""
SELECT
    *
FROM
    df_computer dc
LEFT JOIN
    df_human dh
ON
    dc.file_name = dh.file_name
ORDER BY 
    human_sentiment DESC
"""
)
```

```python
df_sentiment_scores
```

```python
df_sentiment_scores.to_csv('data/computer_human_sentiment_scores.csv', index=False)
```
